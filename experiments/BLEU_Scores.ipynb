In this experiment, the sentence level BLEU scores are calculated between the reference and the different ouptuts. <br /> 
s1 is the reference and s2 is the output generated by the model.

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# Simple and complex sentence pair.

s1 = "John went to the market. On his way back he picked the children."
s2 = "John went to the market and then on his way back he picked the children."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens)
BLEUscore
```

```python
# Missing facts in the output.

s1 = "John went to the market. On his way back he picked the children."
s2 = "John went to the market."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens)
BLEUscore
```

```python
# Repeated facts in the output.

s1 = "John went to the market. On his way back he picked the children."
s2 = "John went to the market. John went to the market. On his way back he picked the children."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens)
BLEUscore
```

```python
# Unsupported facts in the ouptut.

s1 = "John went to the market. On his way back he picked the children."
s2 = "John went to the market. On his way back he saw the children playing."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens)
BLEUscore
```

```python
# Unsupported facts in the output.

s1 = "ISBM numbered 0-7653-0633-6 book is A Fortress of Grey Ice. It has 672 pages."
s2 = "J.V. Jones authored A Fortress of Grey Ice. A Fortress of Grey Ice has 672 pages."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens)
BLEUscore
```

```python
# Unsupported facts in the output.

s1 = "Alan Shepard Alan Shepard Alan Shepard."
s2 = "Alan Shepard is dead. Alan Shepard was a test pilot."
s1_tokens = word_tokenize(s1)
s2_tokens = word_tokenize(s2)
BLEUscore = nltk.translate.bleu_score.sentence_bleu([s1_tokens],s2_tokens, weights=(0.5,0.5))
BLEUscore
```
